```python
class KMeansCustom:
    def __init__ (self, n_clusters, data):
        self.n_clusters = n_clusters #broj grupa, centroida
        self.data = data
        
    def euclidean_distance(a,b):
        return np.sqrt(np.sum(a-b)**2)
    
    def calculate_distances(self):
        distances = np.zeros((self.n_clusters, self.n_clusters))  
        for i in range(self.n_clusters):
            for j in range(i+1, self.n):
                distance = self.euclidean_distance(i,j)
                distances[i, j] = distance
                distances[j, i] = distance
        self.distances = distances
        return self
    """ sets self.clusters """
    def assign_clusters(self):
        clusters = []
        for d in self.data:
            closest_centroid = np.inf
            for c in self.centroids:
                current_centroid = self.euclidean_distance(d,c)
                if current_centroid < closest_centroid:
                    closest_centroid = current_centroid
            clusters.append(closest_centroid)
        self.clusters = clusters
        return self
    
    """ updates centroids """
    def update_centroids(self):
        #for current clusters, get means and set them as new centroids
        unique_labels = np.unique(self.clusters)
        means = []

        for label in unique_labels:
            group = self.data[np.array(self.clusters) == label]
            mean = np.mean(group, axis=0)
            means.append(tuple(mean))  # convert to (x, y)
        self.centroids = means
        return self
    def pprint(self):
        print('check')
        return 'print'
    def max_min_values(self):
        minimum = self.data.min()
        maximum = self.data.max()
        self.centroids = np.random.rand(self.n_clusters, len(self.data[0])) * 100
        print("Max Values:", maximum)
        print("Min Values:", minimum)
        return maximum,minimum
    
    def pipeline(self):
        max_vals, min_vals = self.max_min_values()  # Call the max_min_values method correctly
        print("Max Values:", max_vals)
        print("Min Values:", min_vals)
        self.centroids = np.random.rand(self.n_clusters, len(self.data[0])) * 100
        
        return self
    
    
    
                    
        
        
```
